```python
from bs4 import BeautifulSoup
import requests

url = 'https://news.yahoo.co.jp/articles/bd66addad7564c088a7997b2421a5a2ed2fa83e7'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
article_element1 = soup.select_one('#uamods > .article_body > .sc-zniwbk-0:nth-of-type(1) > p')

article_element2 = soup.select_one('#uamods > .article_body > .sc-zniwbk-0:nth-of-type(2) > p')

article_element3 = soup.select_one('#uamods > .article_body > .sc-zniwbk-0:nth-of-type(3) > p')


article_text1 = article_element1.text
article_text2 = article_element2.text
article_text3 = article_element3.text
print(article_text1 + article_text2 + article_text3)
```
